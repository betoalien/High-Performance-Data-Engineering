# Getting Started with High-Performance DataFrames

This notebook demonstrates the basics of using the hybrid Rust+Python DataFrame engine for interactive data analysis.

## Learning Objectives

By the end of this notebook, you will:
1. Know how to create DataFrames from various sources
2. Understand basic DataFrame operations
3. See performance comparisons with pure Python approaches

```python
# Import the SDK
from pardox import DataFrame, Series, read_csv, read_json
import time

print("PardoX SDK loaded successfully!")
```

## Creating DataFrames

```python
# Method 1: From Python data (list of dicts)
data = [
    {"name": "Alice", "age": 30, "salary": 75000},
    {"name": "Bob", "age": 25, "salary": 65000},
    {"name": "Charlie", "age": 35, "salary": 85000},
    {"name": "Diana", "age": 28, "salary": 70000},
]

df = DataFrame(data)
print(f"Created DataFrame with shape: {df.shape}")
print(f"Columns: {df.columns}")
```

```python
# Method 2: View the data
df.head()
```

## Basic Operations

```python
# Column selection
print("Names:", df["name"].tolist())
```

```python
# Arithmetic operations (vectorized in Rust)
df["salary_doubled"] = df["salary"] * 2
df["salary_with_bonus"] = df["salary"] + 5000

print("Original salaries and calculated columns:")
df[["name", "salary", "salary_doubled", "salary_with_bonus"]].head()
```

```python
# Filtering
high_earners = df[df["salary"] > 70000]
print(f"High earners (salary > $70k): {high_earners.shape[0]} people")
high_earners[["name", "salary"]].head()
```

## Aggregations

```python
# Summary statistics
print(f"Total salary: ${df['salary'].sum():,.0f}")
print(f"Average salary: ${df['salary'].mean():,.0f}")
print(f"Min salary: ${df['salary'].min():,.0f}")
print(f"Max salary: ${df['salary'].max():,.0f}")
```

```python
# GroupBy aggregation
age_groups = df.group_by("age").agg({
    "salary": ["sum", "mean", "count"]
})
age_groups.head()
```

## Performance Comparison

```python
# Generate larger dataset for benchmarking
import random

def generate_data(n_rows):
    return [
        {
            "id": i,
            "value": random.uniform(0, 1000),
            "category": random.choice(["A", "B", "C", "D"]),
            "quantity": random.randint(1, 100)
        }
        for i in range(n_rows)
    ]

# Test with 100k rows
n_rows = 100_000
data = generate_data(n_rows)

print(f"Generated {n_rows:,} rows of test data")
```

```python
# Benchmark: Rust DataFrame creation
start = time.time()
df_rust = DataFrame(data)
rust_time = time.time() - start
print(f"Rust DataFrame creation: {rust_time*1000:.1f}ms")
```

```python
# Benchmark: Pure Python approach
start = time.time()
python_data = data.copy()
python_time = time.time() - start
print(f"Pure Python (just copying list): {python_time*1000:.1f}ms")
```

```python
# Benchmark: Aggregation
# Rust
start = time.time()
rust_result = df_rust.group_by("category").agg({"value": "sum"})
rust_agg_time = time.time() - start

print(f"Rust GroupBy aggregation: {rust_agg_time*1000:.1f}ms")
print(f"Results by category:")
print(rust_result)
```
